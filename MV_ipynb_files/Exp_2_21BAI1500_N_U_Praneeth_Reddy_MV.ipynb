```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

## Min , Max , Mean of the Color Plane

```python
img=cv2.imread("airplane.png")
print("Maximum value of Colour plane :",np.max(img))
print("Minimum value of Colour plane :",np.min(img))
print("Mean value of Colour plane :",round(np.mean(img),4))
```

##BGR

```python
bgr_img=cv2.imread("airplane.png")
```

```python
# Increasing the Brightness of the BGR Image by adding a value to the array
plt.subplot(1,2,1)
plt.imshow(bgr_img)
plt.title("Original BGR Image")
hgh_bgr=bgr_img + 30
plt.subplot(1,2,2)
plt.imshow(hgh_bgr)
plt.title("Brightened Image")
```

```python
# decreasing the Brightness of the BGR Image by adding a value to the array
plt.subplot(1,2,1)
plt.imshow(bgr_img)
plt.title("Original BGR Image")
lw_bgr=bgr_img -25
plt.subplot(1,2,2)
plt.imshow(lw_bgr)
plt.title("Low Brightness BGR Image")
```

```python
# Decreasing the Brightness of the BGR Image by Multiplying a value to the array
plt.subplot(1,2,1)
plt.imshow(bgr_img)
plt.title("Original BGR Image")
lw_bgr=(bgr_img*0.5).astype(int)
plt.subplot(1,2,2)
plt.imshow(lw_bgr)
plt.title("Low Brightness Image")
```

```python
# Increasing the Brightness of the BGR Image by Multiplying a value to the array
plt.subplot(1,2,1)
plt.imshow(bgr_img)
plt.title("Original BGR Image")
hgh_bgr=(bgr_img*1.5).astype(int)
plt.subplot(1,2,2)
plt.imshow(hgh_bgr)
plt.title("Brightened Image")
```
