```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
plt.subplot(1,2,1)
imgbgr = cv2.imread("house2.jpeg")
plt.imshow(imgbgr)
plt.title("Image")

plt.subplot(1,2,2)
gray = cv2.cvtColor(imgbgr,cv2.COLOR_BGR2GRAY)
plt.imshow(gray,'gray')
plt.title("Gray Image")
```

```python
#Horizantal Detection
```

```python
hmask = np.array([[-1,-1,-1],[2,2,2],[-1,-1,-1]])
Horizantal_detect=cv2.filter2D(src=gray, ddepth=-1, kernel=hmask)
```

```python
plt.imshow(Horizantal_detect,"gray")
```

```python
#Vertical Detection
```

```python
vmask = np.array([[-1,2,-1],[-1,2,-1],[-1,2,-1]])
vertical_detect=cv2.filter2D(src=gray, ddepth=-1, kernel=vmask)
```

```python
plt.imshow(vertical_detect,"gray")
```

```python
#diagonal Detection
```

```python
dmask = np.array([[2,-1,-1],[-1,2,-1],[-1,-1,2]])
diagonal_detect1=cv2.filter2D(src=gray, ddepth=-1, kernel=dmask)
```

```python
plt.imshow(diagonal_detect1,"gray")
```

```python
# Roberts Kernel
rmask_x = np.array([[-1,0],[0,1]])
rmask_y = np.array([[0,-1],[1,0]])

img = cv2.filter2D(src=gray,ddepth=-1,kernel=rmask_x)
img2 = cv2.filter2D(src=gray,ddepth=-1,kernel=rmask_y)

plt.subplot(1,2,1)
plt.imshow(img,"gray")
plt.title("Roberts Gx")
plt.subplot(1,2,2)
plt.imshow(img2,"gray")
plt.title("Roberts Gy")
```

```python
#sobel Kernel
```

```python
# Sobel Kernel
smask_x = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
smask_y= np.array([[-1,0,1],[-2,0,2],[-1,0,1]])

img = cv2.filter2D(src=gray,ddepth=-1,kernel=smask_x)
img2 = cv2.filter2D(src=gray,ddepth=-1,kernel=smask_y)

plt.subplot(1,2,1)
plt.imshow(img,"gray")
plt.title("Sobel Gx")
plt.subplot(1,2,2)
plt.imshow(img2,"gray")
plt.title("Sobel Gy")
```

```python
# Prewitt Kernel
pmask_x = np.array([[-1,-1,-1],[0,0,0],[1,1,1]])
pmask_y = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])

img = cv2.filter2D(src=gray,ddepth=-1,kernel=pmask_x)
img2 = cv2.filter2D(src=gray,ddepth=-1,kernel=pmask_y)

plt.subplot(1,2,1)
plt.imshow(img,"gray")
plt.title("Prewitt Gx")
plt.subplot(1,2,2)
plt.imshow(img2,"gray")
plt.title("Prewitt Gy")
```

```python
# Median Filtering
def median_filter(image, filter_size):
    result = np.zeros_like(image)
    indexer = filter_size // 2
    for i in range(indexer, image.shape[0] - indexer):
        for j in range(indexer, image.shape[1] - indexer):
            window = image[i - indexer : i + indexer + 1, j - indexer : j + indexer + 1]
            result[i, j] = np.median(window)
    return result

median = median_filter(noisy,5)
plt.imshow(median)
```

```python
# Sobel on the Median Filtered given image
img = cv2.filter2D(src=median,ddepth=-1,kernel=smask_x)
img2 = cv2.filter2D(src=median,ddepth=-1,kernel=smask_y)

plt.subplot(1,2,1)
plt.imshow(img,"gray")
plt.title("Sobel Gx")

plt.subplot(1,2,2)
plt.imshow(img2,"gray")
plt.title("Sobel Gy")
```
