```python
#import Libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#BGR image
imgbgr = cv2.imread("deer.jpg")

#gray Image
gray =cv2.cvtColor(imgbgr, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(10,8))
plt.subplot(1,2,1)
plt.imshow(imgbgr)
plt.title("BGR Image")

plt.subplot(1,2,2)
gray_image = cv2.cvtColor(imgbgr, cv2.COLOR_BGR2GRAY)
plt.imshow(gray,'gray')
plt.title("Gray Image")
```

### 1.a)Image Negative

```python
# gray negative Image
gray_neg = abs(255-gray)

#plotting
plt.figure(figsize=(10,8))
plt.subplot(1, 2, 1,)
plt.title("Gray Image")
plt.imshow(gray,'gray')

plt.subplot(1, 2, 2)
plt.title("Negative Gray Image")
plt.imshow(gray_neg,'gray')
```

### 1.b)Log transform

```python
# BGR log Transform
a1 = imgbgr/255
log_trans_bgr = 1* np.log(1 + a1)

#plotting for BGR Image log transform
plt.figure(figsize=(10,8))
plt.subplot(1, 2, 1,)
plt.title("Noraml BGR Image")
plt.imshow(imgbgr)
plt.subplot(1, 2, 2)
plt.title("Log Transformed BGR Image")
plt.imshow(log_trans_bgr)

#gray log transform
a = gray/255
log_trans = 3 * np.log(1 + a)

#plotting for gray log transform
plt.figure(figsize=(10,8))
plt.subplot(1, 2, 1,)
plt.title("Noraml Gray Image")
plt.imshow(gray,'gray')

plt.subplot(1, 2, 2)
plt.title("Log Transformed Gray Image")
plt.imshow(log_trans,'gray')
```

### 1.c) Gamma Correction

```python
img_gamma=np.array([((i / 255)**(2.2))* 255 for i in gray]).astype('int')

#plotting images
plt.figure(figsize=(10,8))
plt.subplot(1,2,1)
plt.imshow(gray,"gray")
plt.title("Gray Image")

plt.subplot(1,2,2)
plt.imshow(img_gamma,'gray')
plt.title("Gamma Transformed Gray Image")
```

### 2)Threshold check

```python
# Threshold check
```

```python
thr=[]
for i in gray:
    temp=[]
    for j in i :
        if j < 120:
            temp.append(0)
        else:
            temp.append(1)
    thr.append(temp)

#plotting images
plt.figure(figsize=(10,8))
plt.subplot(1,2,1)
plt.imshow(gray,"gray")
plt.title("Gray Image")
plt.subplot(1,2,2)
plt.imshow(thr,"gray")
plt.title("Threshold Binary Gray Image")
```

### 3)Contrast Stretching

```python
contrast_gray_img= []
for i in gray:
  temp = []
  for j in i:
    if(j<80):
      temp.append(j)
    elif (j<160):
      temp.append(j*7+30)
    else:
      temp.append(j*13-10)
  contrast_gray_img.append(temp)

plt.figure(figsize=(10,8))
plt.subplot(1,2,1)
plt.imshow(gray,"gray")
plt.title("Gray Image")

plt.subplot(1,2,2)
plt.imshow(contrast_gray_img,"gray")
plt.title("Contrast Streched Gray Image")
```

```python
#Plotting Histogram
flat_image = gray.flatten()
flat_processed_image = np.array(contrast_gray_img).flatten()

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(flat_image, bins=256,range=[0, 255], color='blue', alpha=0.7)
plt.title('Histogram of Original Gray Image')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(flat_processed_image, bins=256,range=[0, 255], color='red', alpha=0.7)
plt.title('Histogram of Contrast Stretched Image')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
```
