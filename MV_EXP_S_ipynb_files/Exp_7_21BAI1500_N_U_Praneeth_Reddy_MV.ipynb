```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img= cv2.imread('image.png')
plt.title('Original Image')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
mask = np.zeros(img.shape[:2], np.uint8)
backgroundModel = np.zeros((1, 65), np.float64)
foregroundModel = np.zeros((1, 65), np.float64)
```

```python
rectangle = (400, 50,420 , 650)
cv2.grabCut(img, mask, rectangle,backgroundModel, foregroundModel,5, cv2.GC_INIT_WITH_RECT)
```

```python
mask2 = np.where((mask == 2)|(mask == 0), 0, 1).astype('uint8')
image_segmented = img * mask2[:, :, np.newaxis]
```

```python
plt.figure(figsize=(12,18))
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

plt.subplot(1,2,2)
plt.title('BackGround Removed Image')
plt.imshow(cv2.cvtColor(image_segmented, cv2.COLOR_BGR2RGB))
```
