```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
bgr_fruits =cv2.imread("fruits.png")
if(bgr_fruits.ndim == 3 ):
  print("Image Type : Colour Image")
else :
  print ("Image Type : Greyscale Image")
```

```python
shape =bgr_fruits.shape
print("Image Resolution is :",f"{shape[0]} x {shape[1]} pixels ")
```

```python
#BGR Image of Fruits
plt.imshow(bgr_fruits)
plt.title ("BGR Fruits ")
```

```python
print ("Maximum value of the array for the BGR Image :",np.max(bgr_fruits))
print ("Manimum value of the array for the BGR Image :",np.min(bgr_fruits))
```

```python
# BGR to RGB Convertion
rgb_fruits=cv2.cvtColor(bgr_fruits,cv2.COLOR_BGR2RGB)
plt.imshow(rgb_fruits)
plt.title("RGB Fruits")
```

```python
print ("Maximum value of the array for the RGB Image :",np.max(rgb_fruits))
print ("Manimum value of the array for the RGB Image :",np.min(rgb_fruits))
```

```python
hsv_fruits = cv2.cvtColor(bgr_fruits,cv2.COLOR_BGR2HSV)[:,:,2]
plt.imshow(hsv_fruits)
plt.title("HSV Fruits")
```

```python
print ("Maximum value of the array for the HSV Image :",np.max(hsv_fruits))
print ("Manimum value of the array for the HSV Image :",np.min(hsv_fruits))
```

```python
# Grey Image of the Image
grey_image_fruits =cv2.cvtColor(bgr_fruits, cv2.COLOR_BGR2GRAY)
plt.imshow(grey_image_fruits, "gray")
plt.title("Grey Image of Fruits")
```

```python
print ("Maximum value of the array for the Gray Image :",np.max(grey_image_fruits))
print ("Manimum value of the array for the Gray Image :",np.min(grey_image_fruits))
```

```python
fruits_crop = bgr_fruits[100:300]
plt.imshow(fruits_crop)
plt.title("Cropped Image of Fruits")
```
